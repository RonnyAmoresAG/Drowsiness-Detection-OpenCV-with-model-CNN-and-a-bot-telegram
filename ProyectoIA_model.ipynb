```python
# ESTO FUNCIÓN obtener y preprocesar los datos de entrenamiento
#El ruido en la data de imágenes que usas en tu modelo podría indicar que hay información no deseada en las imágenes, como elementos que no son relevantes para el problema 
#de clasificación o etiquetado que estás tratando de resolver.

#La data de imágenes que utilizas en tu modelo de redes neuronales convolutionales (CNN) es una data no lineal,
# ya que las imágenes tienen una estructura espacial y una distribución de pixeles que no puede ser representada por una función lineal simple.

#for yawn and not_yawn. Take only face //Toma para las imágenes de bostezo para que tome la cara de la matriz de pixeles de la imágen de yawn 
#Este fragmento de código define una función llamada face_for_yawn que utiliza la biblioteca de visión por computadora OpenCV para detectar y recortar rostros de imágenes. La función toma dos argumentos opcionales: direc y face_cas_path. 
#El argumento direc es la ruta del directorio que contiene las imágenes y face_cas_path es la ruta del archivo de clasificador de cascada de Haar que se utiliza para detectar rostros.

#La función recorre las imágenes en cada categoría ("yawn" y "no_yawn") y utiliza el clasificador de cascada de Haar para detectar rostros en cada imagen. Luego, dibuja un rectángulo alrededor de cada rostro detectado 
#y recorta el rostro para obtener una imagen más pequeña. Finalmente, la función redimensiona la imagen recortada a un tamaño específico (145x145 píxeles) y agrega la imagen y la categoría a una lista denominada yaw_no. 
#La función devuelve esta lista al final.

#El archivo haarcascade_frontalface_default.xml contiene una cascada de clasificadores entrenados para detectar rostros humanos en imágenes.
#La cascada de Haar es un tipo de clasificador de objetos basado en el análisis de características de imagen. Se utiliza a menudo en el procesamiento de imágenes y el aprendizaje automático para detectar y clasificar objetos en imágenes.
#el hararcascade es de configuraiciñon generica para detectar cualquier tipo de opjeto de una imagen el otro el caarcascade frontal face te detecta el rostro frontal en una imagen 
#Usa bloques que segmenta dentro de la matriz de imágenes y compara cada bloque donde asocia caracterisitcas y determina objetos. 
def face_for_yawn(direc="/content/drive/My Drive/Proyecto/dataset_new/train", face_cas_path="/content/drive/My Drive/Proyecto/prediction-images/haarcascade_frontalface_default.xml"):
    yaw_no = []
    IMG_SIZE = 145 #cantidad de pixeles de la imagen redimensionar que luego sera agregada a la imagen y la categoría a una lista denominada yaw_no, lo mismo con yawn 
    categories = ["yawn", "no_yawn"]#Lista de dos categorias 
    for category in categories:
        path_link = os.path.join(direc, category)
        class_num1 = categories.index(category)#obtiene el índice de cada categoría en la lista categories. El índice de la categoría "yawn" es 0 y el índice de la categoría "no_yawn" es 1.
        print(class_num1)
        for image in os.listdir(path_link):
            image_array = cv2.imread(os.path.join(path_link, image), cv2.IMREAD_COLOR)
            face_cascade = cv2.CascadeClassifier(face_cas_path)
            faces = face_cascade.detectMultiScale(image_array, 1.3, 5)
            for (x, y, w, h) in faces:
                img = cv2.rectangle(image_array, (x, y), (x+w, y+h), (0, 255, 0), 2)
                roi_color = img[y:y+h, x:x+w]
                resized_array = cv2.resize(roi_color, (IMG_SIZE, IMG_SIZE))
                yaw_no.append([resized_array, class_num1])#(addpend=adjunta) si la imagen pertenece a la categoría "yawn", se agrega el par (imagen, 0) a la lista y si la imagen pertenece a la categoría "no_yawn", se agrega el par (imagen, 1) a la lista.
    return yaw_no #la función devuelve la lista yaw_no que contiene las imágenes recortadas y redimensionadas junto con su categoría correspondiente

 #Luego, la función redimensionará las imágenes y agregará las imágenes y las categorías a una lista, que se devolverá y se asignará a la variable yawn_no_yawn.
yawn_no_yawn = face_for_yawn()

#Basicamente lo que hace la función es crear una lista de dos cualidades (yaw now yawn ) y luego les asigno una valor entero a cada categoria. A través del for itera para cada categoria y con el harcascade de la cara, ubica la cara, se dibua 
#un regtanuglo y se corta eso, de manera que la nueva data son esas imágenes Finalmente de los guarda a todas las imagenes con su resepectiva categoria en la variable(yawn_no_yawn) se guarda como dupla (imagen, 0)o (imagen, 1);


#ya que las imagenes las puso en cada categoria que es almacenada en la lista Estos valores indican a qué categoría pertenece cada imagen yawn 0 y no yawn 1
#MUY IMPORTANTE AQUI YA ASIGNAMS VALORES A LAS CATEGORIAS  un número entero asociado a una categoría. la función asigna la etiqueta "0" a las imágenes de la categoría "yawn" y la etiqueta "1" a las imágenes de la categoría "no_yawn".

#La función utiliza la variable categories que es una lista con las dos categorías: "yawn" y "no_yawn", y utiliza el método index() para obtener el índice de cada categoría en la lista categories. 
#El índice de la categoría "yawn" es 0 y el índice de la categoría "no_yawn" es 1.# aqui basicamente preprocesamos los datos y prepararlos para el entrenamiento del modelo.


```

```python
#for closed and open eye
 # La función toma tres argumentos opcionales: dir_path, face_cas y eye_cas. El argumento dir_path es la ruta del directorio que contiene las imágenes, 
 #face_cas es la ruta del archivo de clasificador de cascada de Haar que se utiliza para detectar rostros y eye_cas es la ruta del archivo de clasificador de cascada de Haar que se utiliza para detectar ojos.
def get_data(dir_path="/content/drive/My Drive/Proyecto/dataset_new/train", face_cas="/content/drive/My Drive/Proyecto/prediction-images/haarcascade_frontalface_default.xml", eye_cas="/content/drive/My Drive/Proyecto/prediction-images/haarcascade.xml"):
    labels = ['Closed', 'Open']
    IMG_SIZE = 145
    data = []
    for label in labels:
        path = os.path.join(dir_path, label) 
        class_num = labels.index(label)#obtiene el índice de cada categoría en la lista categories. El índice de la categoría "Closed" es 2 y el índice de la categoría "Open" es 3.
        class_num +=2
        print(class_num)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_COLOR)
                resized_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                data.append([resized_array, class_num])#(imagen, 2) a la lista  si la imagen pertenece a la categoría "Closed", se agrega el par (imagen, 3) pertenece a la categoria "Open"
            except Exception as e:
                print(e)
    return data
  
```

```python
#Cuando se ejecuta la línea data_train = get_data(), se llama a la función get_data sin proporcionar ningún argumento. Esto significa que se utilizarán los valores predeterminados para los argumentos dir_path, face_cas y eye_cas. 
#La función recorrerá las imágenes en las categorías "Closed" y "Open" en el directorio especificado por dir_path y utilizará los archivos de clasificadores de 
#cascada de Haar especificados por face_cas y eye_cas para detectar rostros y ojos en las imágenes. 
#Luego, la función redimensionará las imágenes y agregará las imágenes y las categorías a una lista.
 #Finalmente, la función devolverá esta lista y se asignará a la variable data_train.
data_train = get_data()
```

```python
#extend data and convert array//extiende la data y le convierte en array se unen las dos listas con iamgenes etiquetadas a una varaible ahora numerica y se las convierte en array 
#La función append_data utiliza las funciones face_for_yawn y get_data para obtener dos listas de imágenes y categorías. 
#Luego, utiliza la función extend de Python para combinar las dos listas en una sola lista. 
#Finalmente, la función convierte la lista resultante en un array de NumPy y lo devuelve. Numpy(es una tabla de elementos (generalmente números) todos del mismo tipo,pueden realizar operaciones matemáticas en bloques enteros de datos utilizando una sola función)
#puede con el Numpy acer operaciones de array porque las imagnees en las listas estan en matrices de pixeles???
def append_data():
#     total_data = []
    yaw_no = face_for_yawn()
    data = get_data()
    yaw_no.extend(data) #se unen 
    return np.array(yaw_no)#Basicamente es una array multdimensional con el que se pede hacer operacones matematcas un mejor manejo de los datos, entonces La función devuelve el array numpy resultante que contiene todas las imágenes redimensionadas y etiquetadas de entrenamiento.
```

```python
#new variable to store nueva variable para almacenar esta función
new_data = append_data()
```

```python
#separate label and features #  separada etiqueta y características
#La variable X se inicializa como una lista vacía. Luego, el código itera sobre cada elemento de la lista new_data y extrae las características y la etiqueta de cada elemento.
 #Las características se agregan a la lista X y la etiqueta se agrega a la lista y. 
#Al final del bucle, las listas X y y contienen las características y etiquetas, respectivamente, de todos los elementos de la lista new_data.
X = []  #caracteristicas imágenes que fueron convertidas en arrays en las funciones de arriba, aqui solo van las imagenes
y = [] #son los 0 1 2 3 es decir etiquetas
#uego, el código comienza un bucle for que itera sobre cada elemento de la lista new_data. Cada elemento de new_data es una tupla que contiene las características y la etiqueta de un elemento. El bucle for asigna esta tupla a dos variables: feature y label.

 #En cada iteración del bucle, se agrega feature a la lista X y label a la lista y.

 #Al final del bucle, las listas X y y contienen las características y etiquetas, respectivamente, de todos los elementos de la lista new_data.
for feature, label in new_data:
    X.append(feature)
    y.append(label)
```

```python
#reshape the arrays remodela 
X = np.array(X) #La línea X = np.array(X) convierte la lista X en un array de NumPy.
X = X.reshape(-1, 145, 145, 3) #todas las imagenes tendran un  formato al array, 145 x 145 imagen y 3 de canalces de colores RGB
#El reshape es necesario ya que los modelos de aprendizaje automático, especialmente las redes neuronales, requieren que los datos de entrada tengan una forma específica. 
#Basicaménte es un formato para todas las imagenes que va n a entrar en el modelo, todoso seas de un mismo formato 
```

```python
#LabelBinarizer /skleaarn modulo de librera de apredizaje automotico
from sklearn.preprocessing import LabelBinarizer 
label_bin = LabelBinarizer() #crea un objeto de la clase LabelBinarizer. Esta clase se utiliza para transformar etiquetas de texto o números en etiquetas binarias. 
y = label_bin.fit_transform(y) #el método fit_transform  Este método encuentra los diferentes valores de las etiquetas en la lista y y los transforma en etiquetas binarias. produciría la matriz binaria
#Esto es necesario para que el modelo pueda procesar adecuadamente las etiquetas y realizar la clasificación
```

```python
#label array
y = np.array(y) #con Numpy la matriz binaria pasa a ser un array Numpy 
```

```python
#train test split divide el test y el train de la data 
#importa la función train_test_split del módulo model_selection del paquete sklearn.

#El conjunto de entrenamiento se utiliza para entrenar el modelo de aprendizaje automático, mientras que el conjunto de prueba se utiliza para evaluar cómo bien el modelo ha sido entrenado. 
#Esta división es importante porque queremos asegurarnos de que el modelo sea capaz de generalizar bien a datos que no ha visto antes.
#La función train_test_split toma dos argumentos obligatorios: el array de características (X) y el array de etiquetas (y). Estos arrays deben tener las mismas dimensiones y deben corresponderse. 
from sklearn.model_selection import train_test_split
seed = 40 #una semilla aleatoria para asegurar que los mismos datos siempre se seleccionen para entrenar y probar el modelo.Una semilla aleatoria es un número entero que se utiliza para inicializar el generador de números aleatorios. Si utilizamos la misma semilla cada vez que ejecutamos nuestro código, el generador de números aleatorios siempre generará la misma secuencia de números. Esto es útil cuando queremos reproducir resultados o cuando queremos asegurarnos de que los mismos datos siempre se seleccionen para entrenar y probar el modelo.
test_size = 0.30  #establece el tamaño de la muestra,que es la data prueba en genral en un 30%.
#el 70% restante se seleccionará para el conjunto de entrenamiento.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=seed, test_size=test_size)

#La línea X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=seed, test_size=test_size) llama a la función train_test_split y le pasa dos argumentos: el array de características X y el array de etiquetas y.
 #La función train_test_split divide aleatoriamente los datos en dos conjuntos: un conjunto de entrenamiento y un conjunto de prueba. El tamaño del conjunto de prueba se establece con el argumento test_size. 
 #La semilla aleatoria se establece con el argumento random_state. La función devuelve cuatro arrays: el conjunto de entrenamiento de características, el conjunto de prueba de características, el conjunto de entrenamiento de etiquetas y 
 #el conjunto de prueba de etiquetas.
 #Estos arrays se asignan a las variables X_train, X_test, y_train y y_test, respectivamente. El argumento random_state establece una semilla aleatoria para garantizar que los mismos datos siempre se seleccionen para entrenar y probar
```

```python
#Data Augmentation/ aumento de datos
#La línea de código train_generator = ImageDataGenerator(rescale=1/255, zoom_range=0.2, horizontal_flip=True, rotation_range=30) crea una instancia del objeto ImageDataGenerator con ciertos parámetros. Esta clase se utiliza para realizar una augmentation de datos, es decir, para crear nuevas versiones de las imágenes existentes en el conjunto de datos de entrenamiento. La idea es que esto pueda mejorar el rendimiento del modelo al proporcionarle más ejemplos para aprender.

#Los parámetros especificados en este caso son:

#rescale: establece el factor de escala para reescalar los valores de píxeles de las imágenes. En este caso, se divide por 255 para ajustar los valores a un rango entre 0 y 1.

#zoom_range: establece el rango para el aumento de la imagen. En este caso, se establece en 0.2, lo que significa que las imágenes se ampliarán en un 20% aleatorio.

#horizontal_flip: establece si se deben volver a girar horizontalmente las imágenes. En este caso, se establece en True, lo que significa que se girarán al azar.

#rotation_range: establece el rango de rotación aleatoria de las imágenes. En este caso, se establece en 30, lo que significa que las imágenes se rotarán al azar en un ángulo

#Con esto genera nueva data y entrena ,es como un generador de entrenamiento
train_generator = ImageDataGenerator(rescale=1/255, zoom_range=0.2, horizontal_flip=True, rotation_range=30) #define para el train, entrena con mayor datos 
test_generator = ImageDataGenerator(rescale=1/255)  #define para el test  

#La función flow() toma como argumentos el conjunto de datos y las etiquetas correspondientes, y crea un iterador que devuelve los datos transformados de manera batch. El argumento shuffle se utiliza para mezclar los datos antes de devolverlos en cada iteración.
 #En este caso se deja False ya que se quiere mantener el orden para comparar con las etiquetas.

train_generator = train_generator.flow(np.array(X_train), y_train, shuffle=False) #aplica las transformaciones del primer objeto ImageDataGenerator al conjunto de entrenamiento y crea un iterador con ellas.
test_generator = test_generator.flow(np.array(X_test), y_test, shuffle=False)#aplica las transformaciones del segundo objeto ImageDataGenerator al conjunto de prueba y crea otro iterador con ellas.
#"shuffle"(barajar) es un parámetro que se utiliza para mezclar los datos de entrada antes de utilizarlos para el entrenamiento. se asegura que los datos se recorran en el mismo orden en que se encuentran en el conjunto de entrenamiento.
#de manera que si es false, los datos no se mezclarán. 
#La data generada por los objetos ImageDataGenerator es nueva data, ya que se basa en el conjunto de entrenamiento original pero se aplican transformaciones aleatorias como escalado, rotación, giro y zoom. 
#Esto aumenta la variedad de datos disponibles para entrenar el modelo y ayuda a mejorar su capacidad de generalización.o.

#Sin embargo, las transformaciones se aplican a las imágenes existentes, no se generan nuevas imágenes. Esto ayuda a reducir el sesgo en el conjunto de entrenamiento y aumenta la robustez del modelo. Damos variedad a la data 
#un sesgo se refiere a una tendencia en el modelo para hacer predicciones incorrectas para ciertos grupos de datos.

#El sesgo se evita con la aleteoridad, de manera que ocurriria sesgo si aplicamos estas trasnformaciones con cierto grupo de data y no le damos una aleteoridad. 
#AUMENTA DATA, AUMENTA DATOS, NO AUMENTA IMAGENES, AUMENTA VARIEDAD ALEATORIA. AUMENTA LA CANTIDAD DE DATOS, NO DE IMAGENES
```

```python
#Model 
#¿Quñe es un modelo secuecial? secuencia lineal de capas. Cada capa recibe la salida de la capa anterior como entrada
model = Sequential()
 #ya que estás trabajando con imágenes, es recomendable utilizar una arquitectura de red neuronal secuencial, como una red neuronal convolucional (CNN) 
#para aprovechar las características espaciales de las imágenes. 
#si está trabajando con imágenes, un modelo secuencial de Keras probablemente sea una buena opción para construir una CNN.

model.add(Conv2D(256, (3, 3), activation="relu", input_shape=X_train.shape[1:]))# 256filtros en imagnes, Cada filtro se comporta como una neurona individua,cada uno de ellos se activa (o no) en función de la intensidad de los píxeles a los que está conectado.l 
model.add(MaxPooling2D(2, 2)) #Agrupación Máxima en ventanas, maxpooling reducir la dimensionalidad de las características.

model.add(Conv2D(128, (3, 3), activation="relu")) # 128filtros en imagnes  #relu función(no lineal)de activación ReLU, f(x) = max(0, x). Esto significa que si x es mayor que 0(+), se devuelve x, pero si x es menor o igual a 0(- o 0), se devuelve 0.  
model.add(MaxPooling2D(2, 2))
#Utilizamos capas de redes ,
#Sigmoide(funcion no lineal) tiene salida siempre está entre 0 y 1,buena opción para problemas de clasificación binaria
model.add(Conv2D(64, (3, 3), activation="relu"))# 64filtros en imagnes  KERNEL es una matriz de valores que se utiliza para aplicar una transformación a una imagen o un conjunto de imágenes.  
model.add(MaxPooling2D(2, 2))

model.add(Conv2D(32, (3, 3), activation="relu"))# 32filtros en imagnes  
model.add(MaxPooling2D(2, 2))

model.add(Flatten())#capa oculta. La capa Flatten se utiliza para aplanar los tensores (matrices multidimensionales) en una sola dimensión.
model.add(Dropout(0.5))#Porcentaje de nuronas que se van a dormir, con el fin de evitar que memoricen, esta apagadas 50% de las nueronas. Me ayuda a regular/controlar ese Overfitting
#Todos los nodos de una capa estan conectados a los nodos de la capa anterior y sucesivvamente
model.add(Dense(64, activation="relu"))#capa formada por el vlaor que pusimos de 64 nodos 
model.add(Dense(32,activation="relu"))
model.add(Dense(4, activation="softmax"))#esta es la capa de salida que esta formada por un valor de 4 que son los nodos de salida, es decir las 4 categorias que vamos a predecir, no yawn yawn closed open 
#softmax es otro tipo de funcion de activación, es comúnmente utilizada en problemas de clasificación múltiple, toma un vector de n elementos y devuelve un nuevo vector de n elementos, donde cada elemento es una probabilidad, la suma es 1 al final.
model.compile(loss="categorical_crossentropy", metrics=["accuracy"], optimizer="adam")
#metrics con a que evalua el modelo, loss permite meidr el rendimiento del moelo mediante el entrenmiento
#que haya perdida muy baja me beneficia, mientas más baja mejor, el rendimiento es bueno. En accuracy/presicion me beneficia que sea muy alta, pero debo saber que no debe haber overfitting en esta grafica accuracy vs epoch 
#La opción optimizer especifica el algoritmo de optimización que se va a utilizar para actualizar los pesos del modelo durante el entrenamiento. En este caso, se está utilizando Adam.
model.summary() #proporciona un resumen de la arquitectura del modelo. 
```

```python
history = model.fit(train_generator, epochs=100, validation_data=test_generator, shuffle=True, validation_steps=len(test_generator))
```

```python
#history-OVERFITTING
#Podemos tomar en cuenta que si queremos saber si esta memorizando el modelo en vez de aprender, lo hacemos sabiendo que para una memorizacion siempre habrá mayor 
#rendimiento en el el trainning que en el testeo, cuando deberia ser al reves màs rendimiento en el testing para que pueda aprender. 
#Ojo esto significa que para memorizacion habra una separacion entre la curva de entrenamiento(arriba) y testeo(abajo). Esto es Overfitting. Debe estar arriba testeo y abajo training, pero lo mñas juntas posible.

#Hay que tener en cuenta también que el conjunto de entrenamiento y el conjunto de validación pueden tener distribuciones diferentes, es posible que el rendimiento del modelo sea mejor en el conjunto de entrenamiento
 #ya que se ha entrenado con ellos, pero si al generalizar en el conjunto de validación el rendimiento es similar, esto es un buen indicio que el modelo no esta memorizando.
#Si las curvas de precisión del conjunto de entrenamiento y el conjunto de validación están muy cerca entre sí, significa que tu modelo está generalizando bien. 
#Esto significa que el modelo está aprendiendo las características de los datos de entrenamiento y es capaz de aplicarlas correctamente a los datos de validación.

#tambien mediante las graficas de accuracy vemos que incrementa la pressicion y no se queda contante en un bajo porcentaje de precision como (0-80%)
# y tambien en la grafica de loss debe haber una decrementacion en el valor de este, si este esta en valores altos y constante vamos por mal camino 
#el OVERFITING Se ajusta muy bien a los datos de entrenamiento, pero no generaliza bien a datos nuevos.
#Al tener overfit el modelo pierde la capacidad de captrar los patrones en el conjunto de entrenamiento 
 #y se tiende a fijar más en los detalles quen no aportan a la prediccion ,asi el modelo lo que hace es memorizar, no aprender.
 #Con esto queda claro que quiero evitar el overfitting, porque esto significa que cuando esta entrenando quizas le vaya muy bien al entrenamiento pero cuando este validando/testeando le va mal, por eso surge esta
 #separacion de curvas/graficas , esto significa que en el entrenamiento esta aprendiedno los detalles/memorizando y en la prueba decae por al ser data nunca vista no puede con lo nuevo, puesto que memorizo 
 # Es como ir a un examen habiendo memorizado, letra a letra, el texto de un libro, pero sin comprender el concepto subyacente.
 #entrenar durante demasiado tiempo conduce a overfitting, por eso donde se estabiliza el modelo es donde deberia qeudarse, recortas epocas si  pasa esto
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(accuracy))

plt.plot(epochs, accuracy, "b", label="trainning accuracy")
plt.plot(epochs, val_accuracy, "r", label="validation accuracy")
plt.legend()
plt.show()

plt.plot(epochs, loss, "b", label="trainning loss")
plt.plot(epochs, val_loss, "r", label="validation loss")
plt.legend()
plt.show()
```
