```python
import requests #importa el módulo requests, el cual permite realizar solicitudes HTTP a través de Python.
import cv2  #(OpenCV), que es una librería de visión artificial que proporciona una gran cantidad de herramientas y funciones para el procesamiento de imágenes y vídeos.
import numpy as np #proporciona herramientas para la manipulación y el cálculo de matrices y vectores.
from keras.models import load_model #permite cargar un modelo de red neuronal previamente guardado en un archivo.
from tensorflow.keras.utils import img_to_array #convierte una imágen de entrada en un arreglo numpy.
from playsound import playsound #permite la reproducción de un archivo de sonido
from threading import Thread #permite ejecutar varias tareas en pararalelo en un programa de pyhton.

#parámetros necesarios para el BOT DE TELEGRAM

TOKEN= "" #TOKEN ÚNICO BOT
chat_id = "" #CHAT ID DEL USUARIO, solo cambia si el usuario cambia de número y/o si el usuario elimina su número/cuenta
message ="PELIGRO: CONDUCTOR DURMIENDOSE: PELIGRO" #MENSAJE DE ALERTA

def start_alarm(sound): #La función de la alarma con el argumento sonido
    """Play the alarm sound"""
    playsound('C:\\Users\\Ronny Amores\Desktop\Quinto Semesre EPN\\IA\Proyecto\\ProyectoIA\\Proyecto Final\\data\\alarm.mp3')#ruta de donde esta el archivo de sonido

    #Definimos la función 
    #def enviar_numero(codigo):

    # Establecer la URL del servicio web y el número a enviar
    #url = f'http://172.31.107.51:5000/api/v1/enviar_numero/{codigo}'

    # Enviar la solicitud HTTP POST al servicio web
    #datos = requests.post(url)

    #codigo = 0
   # 
    # Manejar la respuesta del servidor
    #if datos.status_code == 200:
     #   respuesta = datos.content.decode()
        #print(f'{respuesta}')
    #else:
     #   print('Ocurrio un error al enviar el numero. Codigo de estado HTTP: {}'.format(datos.status_code))

#RECORDERIS: CLASIFICADORES CASCADA
#La idea detrás de los clasificadores de cascada es que se divide la imagen en pequeñas secciones y se busca patrones específicos en cada sección.
    
    
classes = ['Closed', 'Open'] #las dos etiquetas de la clase "Cerrado" y "Abierto", lo que vamos a predecir
face_cascade = cv2.CascadeClassifier("C:\\Users\\Ronny Amores\\Desktop\\Quinto Semesre EPN\\IA\Proyecto\\ProyectoIA\\Proyecto Final\\data\\haarcascade_frontalface_default.xml")
left_eye_cascade = cv2.CascadeClassifier("C:\\Users\\Ronny Amores\Desktop\Quinto Semesre EPN\\IA\Proyecto\\ProyectoIA\\Proyecto Final\\data\\haarcascade_lefteye_2splits.xml")
right_eye_cascade = cv2.CascadeClassifier("C:\\Users\\Ronny Amores\Desktop\Quinto Semesre EPN\\IA\Proyecto\\ProyectoIA\\Proyecto Final\\data\\haarcascade_righteye_2splits.xml")
cap = cv2.VideoCapture(0)#crea un objeto de caputra de video de la camara web principal "0=principal, 1=secundaria"
model = load_model("C:\\Users\\Ronny Amores\\Desktop\\Quinto Semesre EPN\\IA\\Proyecto\\ProyectoIA\\Proyecto Final\\SlumberSensor.h5")
#C:\\Users\\Ronny Amores\\Downloads\\SlumberSensor.h5
count = 0 #contador en 0
alarm_on = False #banderita en falso 
alarm_sound = "C:\\Users\\Ronny Amores\\Desktop\\Quinto Semesre EPN\\IA\\Proyecto\\Pruebas IIB\\driver_drowsiness_system_CNN-main\\driver_drowsiness_system_CNN-main\\data\\alarm.mp3"
status1 = '' #inicializa el status1 que sera del ojo izq con una cadena vacía, luego tendra valores de "2" o "3"
status2 = '' #inicializa el status2 que sera del ojo der con una cadena vacía, luego tendra valores de "2" o "3"

while True: #bucle infinito hasta que lo detenga manualmente
    #En cada iteración del bucle, esta línea lee un fotograma de video del objeto cap, que se define anteriormente como
    #la captura de video de la cámara. El primer valor devuelto por la función cap.read() se ignora mediante el uso de _ y el segundo valor se asigna a frame.
    _, frame = cap.read()
    height = frame.shape[0]#extrae la altura de la imagen de video caputarada en el fotograma actual
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)#convierte la imagen de video capturada en escala de grises
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)#detecta las caras con la imagen de video en gray(1.3 y 5 son parametros de escala y los vecino minimos respectivamente, que afectan)
    #la precision y velocidad de la detección de caras
    for (x, y, w, h) in faces: #recorre todas la caras detectadas en la imagen de video  extrae las cordenadas del rectángulo que rodea cada cara
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 1) #dibuja el rectángulo alrededor de la cada detectada en la imagen del video 
        roi_gray = gray[y:y+h, x:x+w] #extrae la región de interes (ROI) de la cara detectada en escala de grises y almacena en roi_gray
        roi_color = frame[y:y+h, x:x+w]  #extrae la ROI de la cara detectada en la imagen del video orgiinal a color
        left_eye = left_eye_cascade.detectMultiScale(roi_gray)#detecta el ojo izquierdo en la ROI de la cara a escala gris(roi_gray) utilizando el clasificador de cascada Haar del ojo izq
        right_eye = right_eye_cascade.detectMultiScale(roi_gray)#detecta el ojo derecho en la ROI de la cara a escala gris(roi_gray) utilizando el clasificador de cascada Haar del ojo der
        for (x1, y1, w1, h1) in left_eye:#itera través de cada detección de ojo izquierdo
            cv2.rectangle(roi_color, (x1, y1), (x1 + w1, y1 + h1), (0, 255, 0), 1)#se dibuja un rectangulo verde alderedor de este ojo en elimagen original "frame", el roi color es la región de interés 
            eye1 = roi_color[y1:y1+h1, x1:x1+w1]#Se extrae el ojo izquierdo actual de la ROI, utilizando sus coordenadas x, y, w y h.
            eye1 = cv2.resize(eye1, (145, 145)) #tamaño del ojo se ajusta a (145,145) para la entrada de la imágen para el modelo(SlumberSensor.h5)
            eye1 = eye1.astype('float') / 255.0 #valores de pixeles de la imágen se normalizan dividiéndolos por 255.0
            eye1 = img_to_array(eye1) #la imágen del ojo se convierte en un arreglo Numpy
            eye1 = np.expand_dims(eye1, axis=0) #Se agrega una dimensión al arreglo NumPy, lo que significa que la imagen del ojo se trata como un lote de tamaño 1.
            pred1 = model.predict(eye1) #el modelo hace una predicción en la imagen del ojo actual
            status1=np.argmax(pred1) # se seleccion la clase predicha 2 o 3, a partir del resultado de la predicción con la función argmax
            # La función np.argmax() encuentra el índice de la clase con la mayor probabilidad y lo almacena en la variable status1.
            break

        for (x2, y2, w2, h2) in right_eye:
            cv2.rectangle(roi_color, (x2, y2), (x2 + w2, y2 + h2), (0, 255, 0), 1)
            eye2 = roi_color[y2:y2 + h2, x2:x2 + w2]
            eye2 = cv2.resize(eye2, (145, 145))
            eye2 = eye2.astype('float') / 255.0
            eye2 = img_to_array(eye2)
            eye2 = np.expand_dims(eye2, axis=0)
            pred2 = model.predict(eye2)
            status2=np.argmax(pred2)
            break

        #codigo = 0
        #enviar_numero(codigo)

        
        if status1 == 2 and status2 == 2: #2-Closed- Etiqueta de predicción: ambos ojos están cerrados
            count += 1 #empiezael incremento de la variable
            cv2.putText(frame, "Eyes Closed, Frame count: " + str(count), (10, 30), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 255), 1)
            # if eyes are closed for 3 consecutive frames, start the alarm
            if count >=3 :#si lleva 3 frames consecutivos con los ojos cerrados , muesra el mensaje de alerta
                cv2.putText(frame, "Drowsiness Alert!!!", (100, height-20), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 255), 2)
                url = f"https://api.telegram.org/bot{TOKEN}/sendMessage?chat_id={chat_id}&text={message}"#url del api de telegram para mandar un mensaje con parámetros de token, mensaje y el id del chat
                cv2.imwrite('alert.jpg',frame)#Realiza el guardado del frame acutal(alerta de somnolencia y la gurda en la carpeta local)
                caption='POR FAVOR CONTACTESE CON EL CONDUCTOR LO MÁS ANTES POSIBLE'#mensaje o pie de imágen a enviar
                files = {'photo':open('C:\\Users\\Ronny Amores\\PROYECTO\\alert.jpg','rb')}#ruta del archivo a enviar
                url = f"https://api.telegram.org/bot{TOKEN}/sendPhoto?chat_id={chat_id}&caption={caption}" 
                respuesta = requests.post(url,files=files)
                if not alarm_on:#si la alarma no esta encencida, se la inciia en un nuevo hilo
                    alarm_on = True
                    t = Thread(target=start_alarm, args=(alarm_sound,))
                    t.daemon = True
                    t.start()
                    #codigo = 1
                    #enviar_numero(codigo)
        else:
            cv2.putText(frame, "Eyes Open", (10, 30), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 1)
            count = 0
            alarm_on = False

    cv2.imshow("Drowsiness Detector", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()#Libera la caputura de video. Cieera el archivo de video o la cámara web
cv2.destroyAllWindows()#Cierra todas las ventas abiertas por la aplicaci
```
